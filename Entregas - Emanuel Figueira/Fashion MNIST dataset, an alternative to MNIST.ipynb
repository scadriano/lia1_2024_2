Esse trecho de código define uma função que visualiza um conjunto de imagens junto com seus rótulos (classes).

```python
import matplotlib.pyplot as plt

def visualiza_imagens(images, labels, nomes_classes):
    plt.figure(figsize=(10, 10))
    for i in range(25):  # Visualiza 25 imagens
        plt.subplot(5, 5, i + 1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(images[i], cmap=plt.cm.binary)
        plt.xlabel(nomes_classes[labels[i]])  # Acesse diretamente
    plt.show()
```

Esse trecho de código define uma lista que contém os nomes das classes correspondentes às imagens do conjunto de dados Fashion MNIST.

```python
# Classes das imagens
nomes_classes = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

Essa função visualiza uma amostra de 25 imagens do conjunto de dados de treinamento (Fashion MNIST) junto com os rótulos correspondentes.

```python
visualiza_imagens(imagens_treino, labels_treino, nomes_classes)
```

Esse trecho de código normaliza os valores dos pixels das imagens, o que é uma etapa importante no pré-processamento de dados para treinamento de modelos de aprendizado de máquina.

```python
# Normaliza os valores dos pixels para que os dados fiquem na mesma escala
imagens_treino = imagens_treino / 255.0
imagens_teste = imagens_teste / 255.0
```

Esse trecho de código define a arquitetura de um modelo de rede neural convolucional (CNN) usando a API Keras do TensorFlow.

```python
# Arquitetura do Modelo -> Feature Learning

# Cria o objeto de sequência de camadas - do Keras
modelo_lia = models.Sequential()

# Adiciona a primeira camada de convolução e max pooling (camada de entrada)
modelo_lia.add(layers.Conv2D(32, (3, 3), activation = 'relu', input_shape = (28, 28, 3)))
modelo_lia.add(layers.MaxPooling2D((2, 2)))

# Adiciona a segunda camada de convolução e max pooling (camada intermediária)
modelo_lia.add(layers.Conv2D(64, (3, 3), activation = 'relu'))
modelo_lia.add(layers.MaxPooling2D((2, 2)))

# Adiciona a terceira camada de convolução e max pooling (camada intermediária)
modelo_lia.add(layers.Conv2D(64, (3, 3), activation = 'relu'))
modelo_lia.add(layers.MaxPooling2D((2, 2)))
```

Esse trecho de código adiciona camadas de classificação ao modelo que está sendo construido. Essas camadas são responsáveis por converter as características extraídas pelas camadas convolucionais em classes finais, que serão usadas para a previsão.

```python
# Adicionar camadas de classificação - Converte para um vetor - flatten
modelo_lia.add(layers.Flatten())
modelo_lia.add(layers.Dense(64, activation = 'relu'))
modelo_lia.add(layers.Dense(10, activation = 'softmax'))
```

:Esse trecho de código compila o modelo de rede neural que foi definido anteriormente. A compilação é uma etapa essencial antes de treinar o modelo, pois especifica como o modelo será treinado.

```python
# Compilação do modelo
modelo_lia.compile(optimizer = 'adam',
                   loss = 'sparse_categorical_crossentropy',
                   metrics = ['accuracy'])

# adam - é um algoritmo de aprendizagem tipo backpropagation!
# loss - função de erro. Isso se resume a uma otimização função matemática!
# metrics - medir o sucesso!
```

Este trecho de código implementa um modelo de rede neural convolucional para classificar dígitos manuscritos usando o conjunto de dados MNIST.

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras import models, layers
import time

# Carregar dados do MNIST
(imagens_treino, labels_treino), (imagens_teste, labels_teste) = mnist.load_data()

# Redimensionar e normalizar as imagens
imagens_treino = np.expand_dims(imagens_treino, axis=-1)  # Adiciona um canal (28, 28, 1)
imagens_teste = np.expand_dims(imagens_teste, axis=-1)    # Adiciona um canal

# Redimensionar para 32x32
imagens_treino = np.array([np.resize(img, (32, 32)) for img in imagens_treino])
imagens_teste = np.array([np.resize(img, (32, 32)) for img in imagens_teste])

# Normalizar as imagens
imagens_treino = imagens_treino / 255.0
imagens_teste = imagens_teste / 255.0

# Criar o modelo
modelo_lia = models.Sequential()
modelo_lia.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 1)))
modelo_lia.add(layers.MaxPooling2D((2, 2)))
modelo_lia.add(layers.Conv2D(64, (3, 3), activation='relu'))
modelo_lia.add(layers.MaxPooling2D((2, 2)))
modelo_lia.add(layers.Conv2D(64, (3, 3), activation='relu'))
modelo_lia.add(layers.MaxPooling2D((2, 2)))
modelo_lia.add(layers.Flatten())
modelo_lia.add(layers.Dense(64, activation='relu'))
modelo_lia.add(layers.Dense(10, activation='softmax'))

# Compilar o modelo
modelo_lia.compile(optimizer='adam',
                   loss='sparse_categorical_crossentropy',
                   metrics=['accuracy'])

# Executar o treinamento e medir o tempo
start_time = time.time()
history = modelo_lia.fit(imagens_treino,
                          labels_treino,
                          epochs=50,  # Aqui está a alteração
                          validation_data=(imagens_teste, labels_teste))
end_time = time.time()

print(f"Tempo de treinamento: {end_time - start_time:.2f} segundos")

# Plotar a perda e a acurácia
plt.figure(figsize=(12, 4))

# Plotar a perda
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Loss (Treinamento)')
plt.plot(history.history['val_loss'], label='Loss (Validação)')
plt.title('Loss durante o Treinamento')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plotar a acurácia
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Acurácia (Treinamento)')
plt.plot(history.history['val_accuracy'], label='Acurácia (Validação)')
plt.title('Acurácia durante o Treinamento')
plt.xlabel('Epochs')
plt.ylabel('Acurácia')
plt.legend()

plt.show()

```
